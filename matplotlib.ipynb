## Pyplot
It is the term used to refer to the Python library for 2D plotting. It is a part of the Matplotlib library.
\n There are 20 Key Factors everyperson should know about the pyplot
- **Data Point**
    - Data points are the points that you want to plot on the graph, like the x and y coordinates of the points.
- **X axis | Y axis**
    - X axis is the horizontal axis and Y axis is the vertical axis. 
    - X axis is the independent variable and Y axis is the dependent variable.
- **Figure**
    - Figure is the container for the plot. It is the top-level object in the Matplotlib library.
- **Axes**
    - Axes is the container for the plot. It contains the charts data and the labels.
- **Plot**
    - Plot is the method that is used to create a chart. It is used to plot the data points on the chart.
- **Marker**
    - Marker is the symbol that is used to represent the data points on the chart.
- **Line Style**
    - Line style is the style of the line that is used to represent the data points on the chart.
- **Color**
    - Color is the color of the line that is used to represent the data points on the chart.
- **Legend**
    - Legend is the label box which tells us which color is for which thing and which line is for which thing.
- **Label**
    - Label is the label that is used to represent the x and y axis on the chart.
- **Title**
    - Title is the label that is used to represent the title of the chart.
- **Grid**
    - Grid is the line that is used to represent the grid of the chart.
- **Function**
    - Function is the method that is used to create a chart. It is used to plot the data points on the chart.
- **Method**
    - Method is the functions that is used to create a chart. It is used in object oriented programming.
- **Parameters**
    - These are the values which we are gonna pass while creating the plot.
- **Keyword Arguments**
    - Keyword arguments are the things which we pass as the values like kwargs name=value.
- **Object Oriented Programming**
    - Object Oriented Programming is a programming paradigm that is used to create objects that have properties and methods.
    - It is used to create a chart.
- **DPI**
    - DPI is the dots per inch. It is the resolution of the chart.
- **Backend**
    - Backend is the graphics library that is used to create the chart.



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday','Saturday'])
y = np.array([45, 20, 31, 68, 53, 22])
myplot=plt.plot(x,y, color="blue", label="2026 Feb Sales Data", marker="o", linestyle="--", linewidth=2, )
for i in range(len(x)):
    plt.text(
        x[i],              # X-coordinate position
        y[i] + 2,          # Y-coordinate position (offset by +2 so it sits slightly above the dot)
        str(y[i]),         # The text to display (the sales number)
        ha='center',       # Horizontally center the text over the dot
        va='bottom',       # Vertically align the text to sit on top of the point
        color='black',     # Color of the text
        fontsize=10        # Font size of the text
    )
plt.title("Daily Sales Bar")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.xlim(0, 5)
plt.ylim(0, 75)
plt.xticks([0,1,2,3,4,5], ['Mo', 'Tu', 'We', 'Th', 'Fr','Sa'])
plt.legend(loc='upper left', )
# adding the grid
plt.grid(color="silver", linestyle=":", linewidth=1,)
plt.show()
```

## Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

continents = np.array(['North','South','East','West'])
people = np.array([1000,2000,2000,3000])
# mypie = plt.pie(people, labels=continents, autopct='%1.1f%%', shadow=True, startangle=90)
mypie = plt.pie(people, labels=continents, autopct='%1.1f%%', colors=['gold','skyblue','lightgreen','coral'])
plt.axis('equal')
plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
scores = [40, 42, 43, 45, 50, 54, 55, 58, 60, 60, 62, 65, 67, 70, 72, 73, 75, 78, 80, 82]
plt.hist(scores, bins=[40, 50, 60, 70, 80, 90], color='purple', edgecolor='black')
plt.title('Scores')
plt.xlabel('Scores')
plt.ylabel('Number of Students')
plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.show()
```

## Subplot Functions
Subplots functions are used to create multiple charts in a single figure. They are useful for comparing different data sets or for creating a grid of plots.

 

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,30,20,50]

plt.figure(figsize=(15, 7))

plt.subplot(1,2,1)
plt.plot(x,y,label='Line Chart', marker='o')
plt.title('Line Chat')
plt.legend()

plt.subplot(1,2,2)
plt.bar(x,y,color='skyblue',label='Bar Chart')
plt.title('Bar Chat')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Industrial Way to make the subplots
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,30,20,50]

fig,ax = plt.subplots(1,2, figsize=(15,7))

ax[0].plot(x,y, marker='o')
ax[0].set_title('Line Chart')

bartitle=ax[1].bar(x,y, color='green')
plt.bar_label(bartitle, padding=3)
ax[1].set_title('Bar Chart')

fig.suptitle('My First Figure')

plt.tight_layout()
plt.show()
```

## Saving Visualizations


Saving visualizations is a common task in data science. Matplotlib provides several methods to save figures. The most common method is to use the `savefig` method. This method takes a filename as an argument and saves the figure to that file. For example, as you can see in the image above.

```python
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4,5]
y=[10,40,30,20,50]

plt.plot(x,y,color='green', linewidth=2, linestyle='--', marker='o', markersize=10)
plt.xlabel('Values of the X')
plt.ylabel('Values of the Y')
plt.grid(True)
plt.title('Line Plot')
plt.ylim(0,60)
plt.xlim(0,6)  
plt.savefig('lineplot.png', dpi=300, bbox_inches='tight')    
plt.show()
```
